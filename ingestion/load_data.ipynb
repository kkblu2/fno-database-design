```python
import pandas as pd
from sqlalchemy import create_engine

# Load CSV
df = pd.read_csv(r"E:\fno-database-design\data\3mfanddo.csv")

# PostgreSQL connection
engine = create_engine(
    "postgresql://postgres:546@localhost:5432/fno_db"
)

# Load into staging table
df.to_sql(
    "staging_fno",
    engine,
    if_exists="replace",
    index=False,
    chunksize=100000
)

print("CSV successfully loaded into staging_fno")
```
